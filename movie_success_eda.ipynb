# 영화 '왕과 사는 남자' 천만 달성 성공 요인 분석 (EDA)

이 노트북은 영화 '왕과 사는 남자'의 흥행 성공 요인을 분석하기 위해 수집된 데이터를 탐색하고 시각화합니다.

## 1. 데이터 로드 및 확인

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import koreanize_matplotlib

# 통합 데이터 로드
df_master = pd.read_csv('../data/processed/movie_master_dataset.csv')
df_daily = pd.read_csv('../data/processed/boxoffice_daily_dataset.csv')

display(df_master.head())
display(df_daily.head())
```

## 2. 기본 통계 및 현황

각 영화별 총 관객수와 TMDB 인기도 등을 비교합니다.

```python
# 영화별 총 관객수 비교 시각화
fig_total = px.bar(df_master, x='movie_title', y='total_audience', 
                   title='영화별 총 관객수 비교', text_auto='.2s', color='movie_title')
fig_total.show()
```

## 3. 흥행 추이 분석

개봉 초기 관객 동원 속도와 누적 관객수 성장 곡선을 분석합니다.

```python
# 일별 관객수 추이
fig_trend = px.line(df_daily, x='date', y='daily_audience', color='movie_title', 
                    title='개봉 후 일별 관객수 추이')
fig_trend.show()

# 누적 관객수 성장
fig_growth = px.line(df_daily, x='date', y='accum_audience', color='movie_title', 
                     title='누적 관객수 성장 곡선')
fig_growth.show()
```

## 4. 유튜브 온라인 화제성 분석

유튜브 트레일러 조회수와 실제 관객수 간의 상관관계를 확인합니다.

```python
# 유튜브 조회수 vs 관객수
fig_yt = px.scatter(df_master, x='total_trailer_views', y='total_audience', 
                    text='movie_title', title='유튜브 트레일러 조회수와 관객수 상관관계', 
                    size='total_audience', labels={'total_trailer_views': '트레일러 조회수', 'total_audience': '총 관객수'})
fig_yt.show()
```

## 5. 분석 결과 요약

1. **압도적인 초기 화제성**: '왕과 사는 남자'는 개봉 초기 압도적인 관객 동원력을 보여주었습니다.
2. **바이럴의 영향**: 유튜브 트레일러 조회수와 관객수 간의 양의 상관관계가 뚜렷하게 나타납니다.
3. **리뷰 평점**: 높은 평점과 인기도가 안정적인 롱런의 기반이 되었습니다.
